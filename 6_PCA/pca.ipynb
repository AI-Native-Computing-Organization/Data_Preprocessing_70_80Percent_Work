```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('sample_dataset.csv')
X=df.iloc[:,0:10].dropna() #first 10 column droping non numbers
X
X.shape
```

```python
np.corrcoef(X,rowvar=False
            )

"""his function computes the Pearson correlation coefficients between all pairs of columns (features) in the dataset X.

rowvar=False specifies that each column represents a variable (feature), while rows represent observations. 
This is the typical structure for PCA"""
```

```python
pca=PCA()#withour setting the dimension
X_reduced=pca.fit_transform(X)
X_reduced
```

# Dimensionality reduction by Variance

```python
pca=PCA(n_components=0.9) #utomatically selects the smallest number of principal components (PCs) that retain ≥90% of 
# the total variance in the data.


# This is a dynamic way to reduce dimensionality while preserving most of the information.
X_red=pca.fit_transform(X)
X_red


```

When pca=PCA() , The variance ration was

array([9.99587622e-01, 3.11760506e-04, 1.00124979e-04, 4.74973818e-07,
       9.18593028e-09, 4.66587250e-09, 2.18596146e-09, 5.87368887e-10,
       3.72328850e-10, 7.03752725e-11])


PC1 explains ~99.96% of the variance (9.99587622e-01).

PC2 explains ~0.031% (3.11760506e-04).

PC3 explains ~0.010% (1.00124979e-04).

The rest are negligible.


# So, what happened here?

PC1 alone explains 99.96%, which is already >90%. Therefore, PCA(n_components=0.9) will only keep PC1 (since it exceeds the 90% threshold).
So, X_red = pca.fit_transform(X) will have 1 column (instead of the original 10).



# Why Only 1 Component?

array([[-116.7579459 ],
       [ 624.59331603],
       [-116.74635847],
       [-364.0650415 ],
       [ -77.40160697],
       [ -91.4486227 ],..............)
Your first PC (PC1) is extremely dominant (99.96% variance).

The remaining PCs contribute almost nothing (~0.04% combined), so they are discarded.



```python
pca=PCA(n_components=9.99687622e-01) #"Keep the smallest number of PCs where the cumulative variance ≥ 99.9687622%."




# This is a dynamic way to reduce dimensionality while preserving most of the information.
X_red=pca.fit_transform(X)
X_red


```

PC1 explains ~99.96% of the variance (9.99587622e-01).

PC2 explains ~0.031% (3.11760506e-04).

PC3 explains ~0.010% (1.00124979e-04).

Here,  

Cumulative variance after PC1: ~99.9587% (below your threshold of 99.9687622%).

Cumulative variance after PC2: 99.9587% + 0.03117% = 99.9899% (above the threshold).

Thus, PCA adds PC2 to meet the stricter threshold (99.9687622% > 99.9587%).

# Selecting the best number of components

```python
import matplotlib.pyplot as plt  # <-- Missing import!
pca=PCA()
pca.fit(X)

x=range(1,len(pca.explained_variance_ratio_)+1)#x is a range of integers representing PC indices (e.g., [1, 2, 3, ..., 10] for 10 features).


y=pca.explained_variance_ratio_ #y contains the variance ratios (e.g., [0.999, 0.0003, ...] from your earlier output).

plt.bar(x,y)

#The first PC (PC_1) will dominate (99.96% in your case), with subsequent PCs having near-zero variance.

# The scree plot helps decide how many PCs to keep:

# If PC1 explains >99% variance, further PCs are likely noise.

# So, PCA(n_components=1) should be perfect
```

```python
plt.bar(x,y)
plt.yscale("log") #Sets the y-axis to a logarithmic scale (base 10).


"""
As other than the first PCA, other PCA's are very smaller. So, we used log on the y axis to make it bigger.


When some values are extremely small (e.g., 0.0001) and others large (e.g., 0.999), a linear 
scale makes small bars invisible.Log scale compresses large differences, making tiny values visible.
"""
```
