```python
import pandas as pd
from  sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
```

```python
df=pd.read_csv("sample_dataset.csv")
df.columns #List down all the columns
```

```python
# creating a new dataset with 10 column and a demo target column
df=df.loc[:,['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension','target']].dropna() #all rows and 10 column from earlier dataset.
#removing NAN using dropna()
df
```

```python
X=df.iloc[:,0:10] # 10 column from the new one
y=df.iloc[:,10] #target column
```

```python
X
```

```python
y
```

# Feature Selection with ANOVA

```python
selector=SelectKBest(f_classif,k=5)
selector.fit(X,y)
selector.transform(X)
```

### What This Does:
f_classif: Measures linear relationship between features and target

### How it works:

Computes ANOVA F-value between each feature and target

Essentially checks if mean feature values differ significantly between classes

Only detects linear relationships

### With your binary target:

Identifies features with the most significant differences in means between class 0 and 1

Good for finding features where one class consistently has higher/lower values

### Output:
Returns a transformed dataset with only the top 5 features

These features show the most significant linear relationships with target

```python
X.columns[selector.get_support()] 

#'mean radius', 'mean perimeter', 'mean area', 'mean concavity','mean concave points' has most correlation with the 'target' column
```

# Feature  selection using mutual information

```python
selector=SelectKBest(mutual_info_classif,k=5)
selector.fit(X,y)
selector.transform(X)
X.columns[selector.get_support()]
```

### mutual_info_classif: 
Measures non-linear dependencies between features and target

### How it works:

Calculates how much information each feature provides about the target class

Can detect any statistical relationship (not just linear)

Works well with both continuous and categorical data

### With your binary target:

Will identify features whose distributions differ most between class 0 and 1

Good for finding complex patterns that distinguish the classes

### Output:
Returns the names of the 5 features with highest mutual information scores

These features have the strongest statistical relationship with your target

# Key Difference

The key difference between SelectKBest with mutual_info_classif and f_classif lies in how they measure feature importance for classification. mutual_info_classif evaluates non-linear statistical dependencies between features and the target, capturing any type of relationship (linear, non-linear, monotonic, or non-monotonic) by measuring how much information each feature provides about the target classes. In contrast, f_classif only detects linear relationships by computing ANOVA F-values, which essentially test whether the mean values of features differ significantly across target classes. While mutual_info_classif is more versatile and can uncover complex patterns, it is computationally heavier. On the other hand, f_classif is faster but limited to identifying features with straightforward mean differences between classes. For your binary classification task (0 vs. 1), mutual_info_classif may select features with intricate distributions that distinguish the classes, whereas f_classif will prioritize features where one class consistently has higher or lower average values.
